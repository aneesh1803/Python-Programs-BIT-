```python
def count_chars(sentence):
    word_count = len(sentence.split())
    digit_count = sum(c.isdigit() for c in sentence)
    uppercase_count = sum(c.isupper() for c in sentence)
    lowercase_count = sum(c.islower() for c in sentence)

    print("\nAnalysis of the sentence:")
    print("Number of words:", word_count)
    print("Number of digits:", digit_count)
    print("Number of uppercase letters:", uppercase_count)
    print("Number of lowercase letters:", lowercase_count)

count_chars("Hello everyone, How are you, I am fine thank you 123")
```

```python
#water jug problem
class State:
    def __init__(self, jug1, jug2):
        self.jug1 = jug1
        self.jug2 = jug2
    def __eq__(self,other):
        return self.jug1 == other.jug1 and self.jug2 == other.jug2
    def __repr__(self):
        return f"{self.jug1},{self.jug2}"
    def __hash__(self):
        return hash((self.jug1,self.jug2))
    

def dfs(current, visited, path):
    if current == goal:
        return True, path+[current]
    visited.add(current)

    #fill jug1
    if current.jug1<jug1cap:
        nextState = State(jug1cap, current.jug2)
        if nextState not in visited:
            found, solutionPath = dfs(nextState, visited, path+[current])
            if found:
                return True, solutionPath

    #fill jug2
    if current.jug2<jug2cap:
        nextState = State(current.jug1, jug2cap)
        if nextState not in visited:
            found, solutionPath = dfs(nextState, visited, path+[current])
            if found:
                return True, solutionPath

    #empty jug 1
    if current.jug1> 0:
        nextState = State(0, current.jug2)
        if nextState not in visited:
            found, solutionPath = dfs(nextState, visited, path+[current])
            if found:
                return True, solutionPath

    #empty jug2
    if current.jug2> 0:
        nextState = State(current.jug1, 0)
        if nextState not in visited:
            found, solutionPath = dfs(nextState, visited, path+[current])
            if found:
                return True, solutionPath

    #pour from 1 to 2
    if current.jug1>0 and current.jug2<jug2cap:
        pourAmount = min(current.jug1, jug2cap-current.jug2)
        nextState = State(current.jug1 - pourAmount, current.jug2+pourAmount)
        if nextState not in visited:
            found, solutionPath = dfs(nextState, visited, path+[current])
            if found:
                return True, solutionPath

    #pour from 2 to 1
    if current.jug2>0 and current.jug1<jug1cap:
        pourAmount = min(current.jug2, jug1cap-current.jug1)
        nextState = State(current.jug1+pourAmount, current.jug2-pourAmount)
        if nextState not in visited:
            found, solutionPath = dfs(nextState, visited, path+[current])
            if found:
                return True, solutionPath

jug1cap = 4
jug2cap = 3
Start = State(4,3)
goal = State(2,0)

visited = set()

found, solutionPath = dfs(Start, visited, [ ])

if found:
    print("Solution path is:")
    for solution in solutionPath:
        print(solution)
else:
    print("Goal state not found")

    
```

```python
from collections import deque
class State:
    def __init__(self, leftMissionaries, leftCannibals, boatPosition):
        self.leftMissionaries = leftMissionaries
        self.leftCannibals = leftCannibals
        self.rightMissionaries = 3 - leftMissionaries
        self.rightCannibals = 3 - leftCannibals
        self.boatPosition = boatPosition
    def __eq__(self, other):
        return (self.leftMissionaries == other.leftMissionaries and self.leftCannibals == other.leftCannibals and self.boatPosition == other.boatPosition)
    def __hash__(self):
        return hash((self.leftMissionaries, self.leftCannibals, self.boatPosition))
    def __repr__(self):
        return f"{self.leftMissionaries}{self.leftCannibals} : {self.rightMissionaries}{self.rightCannibals} - {self.boatPosition}"

def isValid(state):
    if state.leftMissionaries<0 or state.leftCannibals<0 or state.rightMissionaries<0 or state.rightCannibals<0:
        return False
    if (state.leftMissionaries>0 and state.leftMissionaries<state.leftCannibals) or (state.rightMissionaries>0 and state.rightMissionaries<state.rightCannibals):
        return False
    return True

def generateNextStates(current):
    nextStates= []
    if current.boatPosition == 'left':
        for missionaries in range(min(2, current.leftMissionaries)+1):
            for cannibals in range(min(2, current.leftCannibals)+1):
                if missionaries+cannibals<=2 and missionaries+cannibals>=1:
                    nextState = State(current.leftMissionaries - missionaries, current.leftCannibals - cannibals, 'right')
                    if isValid(nextState):
                        nextStates.append(nextState)
    else:
        for missionaries in range (min(2, current.rightMissionaries)+1):
            for cannibals in range (min(2, current.rightCannibals)+1):
                if missionaries+cannibals<=2 and missionaries+cannibals>=1:
                    nextState = State(current.leftMissionaries + missionaries, current.leftCannibals + cannibals, 'left')
                    if isValid(nextState):
                        nextStates.append(nextState)
    return nextStates

def bfs(start, goal):
    visited = set()
    queue = deque([(start,[])])
    while queue:
        current, path = queue.popleft()
        visited.add(current)
        if current == goal:
            return path+[current]
        for states in generateNextStates(current):
            queue.append((states, path+[current]))
    return None

start = State(3,3,'left')
goal = State(0,0,'right')

path = bfs(start, goal)

if path:
    print("path to reach the goal state is: ")
    for state in path:
        print(state)
else:
    print("Goal not reachable")
  
```

```python
def isSafe(board, row, col):
    n = len(board)
    # Check column
    for i in range(n):
        if board[i][col] == 1:
            return False
    
    # Check upper diagonal on left side
    i, j = row, col
    while i >= 0 and j >= 0:
        if board[i][j] == 1:
            return False
        i -= 1
        j -= 1
    
    # Check upper diagonal on right side
    i, j = row, col
    while i >= 0 and j < n:
        if board[i][j] == 1:
            return False
        i -= 1
        j += 1
    
    return True

def solveQueens(board, row, Solutions):
    n = len(board)
    if row >= n:
        Solutions.append([r[:] for r in board])
        return
    
    for col in range(n):
        if isSafe(board, row, col):
            board[row][col] = 1
            solveQueens(board, row + 1, Solutions)
            board[row][col] = 0

def printboard(board):
    n = len(board)
    for i in range(n):
        for j in range(n):
            print(board[i][j], end=' ')
        print()
    print()

def printallsolutions():
    n = 8
    board = [[0] * n for _ in range(n)]
    Solutions = []
    solveQueens(board, 0, Solutions)
    numSolutions = len(Solutions)
    
    if numSolutions == 0:
        print("No solution exists")
        return
    else:
        print("Number of solutions:", numSolutions)
        for idx, solution in enumerate(Solutions, start=1):
            print('Solution:', idx)
            printboard(solution)

printallsolutions()
```
